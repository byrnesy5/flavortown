```python
from collections import Counter
import re

# Define a simple list of English stop words
stop_words = set([
    "the", "and", "to", "of", "a", "in", "it", "is", "that", "for", "on", "with", "as", "this", "i",
    "you", "at", "we", "so", "but", "are", "was", "be", "have", "not", "they", "or", "all", "by",
    "from", "an", "if", "do", "what", "about", "up", "out", "just", "like", "now", "my", "your",
    "me", "no", "there", "how", "when", "then", "it's", "it's", "it's", "it's", "it's", "it's"
])

# Tokenize the script into words and remove stop words
words = [w for w in re.findall(r'\b\w+\b', script.lower()) if w not in stop_words]

# Create bigrams from filtered words
bigrams = zip(words, words[1:])
bigram_counts = Counter(bigrams)

# Print the 20 most common bigrams (excluding stop words)
for bigram, count in bigram_counts.most_common(40):
    print(f"{bigram[0]} {bigram[1]}: {count}")
```
